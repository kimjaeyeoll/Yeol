```python
TowDF=pd.read_csv(FILE3, encoding='cp949')
TowDF
```

```python
TowDF['신고일시']=pd.to_datetime(TowDF['신고일시'])
```

```python
TowDF.info()
```

```python
TowDF.insert(2, 'year', TowDF['신고일시'].dt.year)
TowDF.insert(3, 'month', TowDF['신고일시'].dt.month)
```

```python
pd.DataFrame(TowDF['신고일시'].dt.year==2021)
```

```python
TowDF[TowDF['신고일시'].dt.year==2021] 
```

```python
c=[]
date=[]
for i in range(2021, 2023):
    for j in range(1,13):
        a = TowDF[(TowDF['신고일시'].dt.year==i) & (TowDF['신고일시'].dt.month==j)]['번호'].count()
        if a !=0:
            date2 = str(i) +'.' +str(j)
            date.append(date2)
            c.append(a)
       
        
            

```

```python
numCall=pd.DataFrame({'신고건수':c}, index=date)
numCall
```

```python
plt.rcParams["figure.figsize"]=(13,7)
plt.bar(numCall.index, numCall['신고건수'])   #2022.04부터 견인 유예시간 60분 부여
plt.ylabel('신고건수',fontsize=15)      
plt.title('서울시 전동킥보드 견인 현황', fontsize=20)  
plt.show()
```
